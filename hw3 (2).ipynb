## Problem 1. Fokker-Planck Equation and Survival. (4 points)

An agent starts at coordinate x0 and performs Brownian motion with a diffusion coefficient D on a semi-infinite interval [0, ∞) with an absorbing boundary condition at x = 0. During the process of diffusion this unlucky agent has a tendency to perish with a rate α, i.e. according to the law of radioactive decay.

1. Write down the Fokker-Planck equation for the PDF P(x, t) of an agent position and the boundary conditions. (1 point) 
2. Solve the equation and find P(x, t) analytically. (1 point) 
3. Find the survival probability of an agent as a function of time analytically (keep in mind that it can decease due to the death rate α as well as by being absorbed at the boundary). (1 point) 
4. Write a general expression for the average time of survival and higher raw moments. Make a rough estimate of these for the case α ≫ 1. (1 point)

### Solution

Suppose at first that the agents do not decay whatsoever. Then the Fokker-Planck equation reads out to be $$\frac{\partial P}{\partial t} = D\frac{\partial^2 P}{\partial x^2},$$ with the PDF of interest being $P(x,t)$. Let us now introduce the decay into the problem. We think of it in the following way: imagine an ensemble of agents, each of which follows the PDF of interest. We construct the final PDF by considering the PDF that describes an ensemble member picked at random in the general way. We then introduce decay not as a quantity that impacts individual PDFs, but rather the amount of ensemble members in an exponential way: $N(t) = N(0)e^{-\alpha t}$. Since we can, generally speaking, describe an ensemble member with two PDFs: of it being alive and its intrinsic PDF, we notice that the decay does not concern the intrinsic PDF, while the probability of the ensemble member being alive evolves precisely as $1*e^{-\alpha t}$. Consequently, we study the function $P(x,t)e^{-\alpha t}$.

We must rewrite the Fokker-Planck equation, using the chain rule to differentiate wrt time:    
$$e^{-\alpha t}\frac{\partial P}{\partial t}-\alpha Pe^{-\alpha t} =  De^{-\alpha t}\frac{\partial^2 P}{\partial x^2}$$
$$\frac{\partial P}{\partial t}-\alpha P =  D\frac{\partial^2 P}{\partial x^2}$$
With the boundary conditions of:    
$$P(0, t) = P(\infty, t) = 0$$
$$P(x,0) = \delta(x-x_0)$$

Apply the Fourier transform wrt spatial variable:    
$$\frac{\partial \hat{P}}{\partial t}-\alpha \hat{P} =  -k^2D\hat{P}$$
$$\frac{\partial \hat{P}}{\partial t} =  \hat{P}(\alpha - k^2D)$$
This is quite effortlessly separable
$$\frac{d\hat{P}}{\hat{P}} = dt(\alpha - k^2D)$$
$$\hat{P} = Ce^{(\alpha - k^2D)t}$$
Now to consider the boundary values (where we omitted the decay term because it makes no difference, both because an exponential is never zero and because it vanishes at zero time). 

By the sifting delta property:
$$\hat{P}(k, 0) = e^{-ikx_0} = C$$
$$\hat{P} = e^{(\alpha - k^2D)t-ikx_0}$$

Then in natural-boundary case we get:     

$$P = \frac{1}{2\pi}\int_\mathbb{R} e^{(\alpha - k^2D)t-ikx_0+ikx}dk = \dfrac{\mathrm{e}^{-\frac{x_0^2}{4Dt}+\frac{xx_0}{2Dt}-\frac{x^2}{4Dt}+{\alpha}t}}{2\sqrt{{\pi Dt}}} = \frac{e^{\alpha t - \frac{(x-x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}}$$

Since we require the solution to be instead defined on a positive half-line, we employ the method of images and compose the solution with its mirror image about the origin:     
$$P = \frac{e^{\alpha t - \frac{(x-x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}}-\frac{e^{\alpha t - \frac{(x+x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}}$$
One easily sees that this solution, indeed, yields zero at $x=0$. The only thing left to do to complete the solution is to introduce the exponential decrement:    


Here I had to stop because something obviosuly went wrong: we cannot possibly have a positive exponential with respect to time; that is absurd, since it'd mean the pdf would blow up without the decrement, and the decrement would turn it into the regular solution. Evidently, the correct way is to apply the decrement after the solution is obtained, although it greatly interests me as to why the two approaches are not equivalent. 

In that case, we get:    
$$P = \frac{e^{- \frac{(x-x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}}-\frac{e^{- \frac{(x+x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}}$$
$$P(x,t)e^{-\alpha t} = \frac{e^{-\alpha t- \frac{(x-x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}}-\frac{e^{-\alpha t- \frac{(x+x_0)^2}{4Dt}}}{2\sqrt{\pi Dt}} = \frac{e^{-\alpha t}(e^{- \frac{(x-x_0)^2}{4Dt}} - e^{- \frac{(x+x_0)^2}{4Dt}})}{2\sqrt{\pi Dt}}$$

And can formally express the process thus (thereby rewriting the Fokker-Planck equation as we had initially):    

We now consider the nondecaying ensemble with the FPE of 
$$\frac{\partial P}{\partial t} = D\frac{\partial^2 P}{\partial x^2},$$
but now $P_1 = Pe^{-\alpha t}$, and, therefore, $P = P_1e^{\alpha t}$, whence:    
$$\frac{\partial P_1}{\partial t}+\alpha P_1 =  D\frac{\partial^2 P_1}{\partial x^2}$$
With the boundary conditions of:    
$$P_1(0, t) = P_1(\infty, t) = 0$$
$$P_1(x,0) = \delta(x-x_0)$$
And from this we recover the form on which we settled:    
$$P_1 = \frac{e^{-\alpha t}(e^{- \frac{(x-x_0)^2}{4Dt}} - e^{- \frac{(x+x_0)^2}{4Dt}})}{2\sqrt{\pi Dt}}$$

To recover the survival probability, we marginalize the PDF wrt time:    
$$S(t) = \int_0^\infty P_1(x,t)dx = \int_0^\infty \frac{e^{-\alpha t}(e^{- \frac{(x-x_0)^2}{4Dt}} - e^{- \frac{(x+x_0)^2}{4Dt}})}{2\sqrt{\pi Dt}}dx = \frac{e^{-\alpha t}}{2\sqrt{\pi Dt}}\int_0^\infty (e^{- \frac{(x-x_0)^2}{4Dt}} - e^{- \frac{(x+x_0)^2}{4Dt}}) dx = \frac{e^{-\alpha t}}{2\sqrt{\pi tD}} (-\dfrac{\sqrt{{\pi}}\left(\operatorname{erf}\left(\frac{\sqrt{\frac{1}{Dt}}x}{2}+\frac{x_0}{2D\sqrt{\frac{1}{Dt}}t}\right)-\operatorname{erf}\left(\frac{\sqrt{\frac{1}{Dt}}x}{2}-\frac{x_0}{2D\sqrt{\frac{1}{Dt}}t}\right)\right)}{\sqrt{\frac{1}{Dt}}})|_0^\infty = \frac{-e^{-\alpha t}}{2}((1-1)-(2\operatorname{erf}(\frac{x_0}{2\sqrt{Dt}}))) = e^{-\alpha t}\operatorname{erf}(\frac{x_0}{2\sqrt{Dt}})$$

We can check that this is a valid, if not necessarily true, function, as, given nonnegative parameters, it is nonincreasing.

```python

def randwalkcorr(tau = 0.5, gamma = 2, x0 = 0.5, absorbers = [0,1], steps = 100000, stepsize = 0.001): #gamma is prefactor in correlator of auxiliary white noise
  x = x0
  y = 0
  for i in range(steps):
    y = -y/tau+np.sqrt(2)/tau*random.normalvariate(0,1*stepsize) #as per paper quoted in the lectures
    x+=y
    if x>=absorbers[1] or x<=absorbers[0]:
      return i, (x>=absorbers[1])*'right'+(x<=absorbers[0])*'left'
```

Generally, $$_0D^α_xP(x,t) = \frac{1}{\Gamma(\lfloor\alpha\rfloor+1-\alpha)}(e^x\int_0^x{y^{\lfloor\alpha\rfloor+1-\alpha}e^{-y}dy}+x^{\lfloor\alpha\rfloor+1-\alpha}e^{-x})$$
The floor-plus-one notation is deliberate here, as I've noticed that ceiling would give wrong transition at integeres.

```python
def frd2(a, x):
  b = np.floor(a)+1
  print(b-a+1)
  incomplete_gamma = scipy.special.gammainc(b-a+1, x)*scipy.special.gamma(b-a+1) #since special returns normalized by gamma value, and I don't want to lose anything to book
  #keeping
  return 1/scipy.special.gamma(b-a)*(np.exp(x)*incomplete_gamma+np.exp(-x)*(x**(b-a)))
```

## Problem 4 (Superdiffusion and fractional Fokker-Planck equations, 3 points)

One of the ways to simulate the superdiffusion is to use the overdamped Langevin equation with the white L ́evy noise instead of a usual Gaussian noise, i.e. the random variables are taken from the symmetricα-stable distribution. The family of α-stable distributions is a class of distributions which are produced by a sequence of variables with a diverging second moment (according to the generalised central limit theorem). The symmetric α-stable distribution has a characteristic function φ(k) = exp (−σμ|k|μ), where σ is a measure of a width of the distribution and μ is a power-law exponent. The values from α-stable distributions can be sampled, for instance, according to the equations
$$\xi_\mu = \frac{sin(\mu \Phi)}{cos(\Phi)^{1/\mu}}(\frac{cos((1-\mu)\Phi)}{W})^{1/\mu - 1}$$
Here W is an exponential variable (P(W > w) = exp(−w),w ≥0) and Φ is uniform on [−π/2,π/2] (More details you could find in chambers1976.pdf). The rest of the simulation procedure is identical to the Brownian case (Each jump has the same duration). The correponding discretisation of the equation is, however, slightly different,
$$x_{n+1} = x_n+\xi_\mu(n)(K_\mu\delta t)^{1/\mu}$$
xn are the positions, Kμ is a fractional diffusion coefficient with dimension cmμ/sec and ξμ(n) is the α-stable noise which is delta correlated ⟨ξμ(n)ξμ(n′)⟩ = δ(n −n′), and samples values according to Eq. (2). μ is used here instead of α in the problem one to avoid confusion. This equation allows to perform a simulation of the process called as L ́evy flights.
1. Simulate a trajectories for L ́evy flights in 1D for the starting position x0 = 0, plot them for μ = 0.5; 1; 1.5 and compare them visually with trajectories for the Brownian motion (μ = 2).
2. Plot the probability distribution functions for t = 0.1,1,10,100 for stable exponent μ = 0.5; 1; 1.5 (Reminder. μ = 2 corresponds to the Brownian motion and 0 < μ < 2 to the scale-free L ́evy flights).

```python
rng = np.random.default_rng() #time to update my np rng usage
```

```python
def xi(mu, n = None):
  w = rng.exponential(size = n)
  phi = rng.uniform(low = -np.pi/2, high = np.pi/2, size = n)
  return np.sin(mu*phi)/(np.power(np.cos(phi), 1/mu))*np.power(np.cos(phi*(1-mu))/w, 1/mu-1)
```

```python
def superdiff(mu, steps = 10000, x0 = 0, kmu = 1, steplen = 1):
  x = np.zeros(steps)
  x[0] = x0
  for i in range(steps-1):
    x[i] = x[i-1]+xi(mu)*np.power(kmu*steplen,1/mu)
  return x
```
