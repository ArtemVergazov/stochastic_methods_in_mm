```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 1000
M = 1
```

```python
x = np.linspace(expon.ppf(.01), expon.ppf(.99), num=size)
X = expon.rvs(size=size)
plt.plot(x, expon.pdf(x))
plt.hist(X, bins=30, density=True);
```

```python
u = np.random.uniform(size=size)
```

```python
p = lambda x: np.sqrt(2 / np.pi) * np.exp(-x**2 / 2)
```

```python
X1 = X[u < p(X) / M / expon.pdf(X)]
X1.shape
```

```python
x = np.linspace(0, 10, num=size)
plt.plot(x, p(x))
plt.hist(X1, bins=200, density=True)
plt.xlim(0, X1.max());
```
