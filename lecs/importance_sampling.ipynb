```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = lambda x: 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)
f = lambda x: np.exp(-(x - 4)**2 / 2)
p_a = lambda x: 1 / np.sqrt(np.pi) * np.exp(-(x - 2)**2)
```

```python
x = np.linspace(-10, 10, num=100)
plt.plot(x, p(x), '-', label='$p(x)$')
plt.plot(x, f(x), '-', label='$f(x)$')
plt.plot(x, p_a(x), '-', label='$p_a(x)$')
plt.legend();
```

```python
X = np.random.randn(1000)
avg = (f(X) * p(X) / p_a(X)).mean()
avg
```
